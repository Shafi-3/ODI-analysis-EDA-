```python
# Identify the player with the highest batting average.
df.columns
```

```python
df['Total Runs'] = pd.to_numeric(df['Total Runs'], errors='coerce')
df['Inning Played'] = pd.to_numeric(df['Inning Played'], errors='coerce')
df['Not Out\'s'] = pd.to_numeric(df['Not Out\'s'], errors='coerce')
```

```python
df['Not Out\'s'] = df['Not Out\'s'].fillna(0)
```

```python
df['Inning Played'] = df['Inning Played'].fillna(0)
```

```python
df['Not Out\'s'] = df['Not Out\'s'].fillna(0)
```

```python
df['Times Out'] = df['Inning Played'] - df['Not Out\'s']
```

```python
df['Batting Average'] = df['Total Runs']/df['Times Out']
```
